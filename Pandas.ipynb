<font size=6>Merging the data</font>

<font size=6>To work on multiple files we will <br>
first merge data<br>
then, read that data file</font>

<font size=6>Merging is done on common col.</font>

<font size=6>four types of merging<br>
Inner join - only matching keys are going to be merged other will be left out or not considered or ignored<br>
Outer join - all the matching values will be merged with all the non matching as well<br>
Left join - all the values from left table will be merged<br>
Right join - all the values from right table will be merged</font>

```python
market = pd.read_csv('market_fact.csv')
customer = pd.read_csv('cust_dimen.csv')
product = pd.read_csv('prod_dimen.csv')
```

```python
customer.head()
```

```python
product.head()
```

```python
newdf = pd.merge(market,customer, how= 'inner',on= 'Cust_id')
newdf.head()
```

<font size=6>market : 'Cust_id'<br>
customer : 'Id'<br>
pd.merge(market, customer, how= 'inner', left_on= 'Cust_id, right_on= 'Id')</font>

```python
masterdf = pd.merge(newdf,product, how= 'inner', on= 'Prod_id')
masterdf.head()
```

<font size=6>To save merged data file.</font>

```python
masterdf.to_csv('merged_data.csv', index= False)
```
