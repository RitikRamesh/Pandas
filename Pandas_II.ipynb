<font size = 6>Few options to clean missing data:<br>

2. delete the row<br>
Filling the data(using mean, median, mode)</font>

```python
housing.shape
```

```python
housing.isnull().sum(axis=1)
```

```python
housing.isnull().sum()
```

<font size = 6>Threshold - the max amount with which one can go.</font>

<font size = 5>Deleting a coln</font>

<font size = 4>deletes the data temporary not in the original file.</font>

```python
housing.drop(columns=['BuildingArea','YearBuilt','CouncilArea','Landsize'])
```

<font size = 4>to delete coln in function</font>

```python
housing = housing.drop(columns=['BuildingArea','YearBuilt','CouncilArea','Landsize'])
```

```python
housing.shape
```

<font size = 5>Deleting the row</font>

```python
housing.isnull().sum(axis=1) > 4
```

```python
housing[housing.isnull().sum(axis=1)>4]
```

```python
housing = housing[housing.isnull().sum(axis=1)<=4]
housing.head()
```

```python
housing['Price'].fillna(housing['Price'].mean())
```

```python
housing=housing['Price'].fillna(housing['Price'].median())
housing
```

<font size = 6>Finding duplicate rows and colns</font>

```python
housing[housing.duplicated()]
```

```python
housing=housing.drop(15858, axis=0)
```

```python
housing[housing.duplicated()]
```

<font size = 6>Mention index no. in () in below code</font>

```python
housing = housing.drop_duplicates()
```

```python
housing.head()
```

<font size = 6>Renaming the coln</font>

```python
housing = housing.rename(columns={'bedroom2':'bedroom'})
housing.head()
```
